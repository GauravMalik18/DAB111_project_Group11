```python
import sqlite3 
import csv 

# establish a connection
conn = sqlite3.connect('supermarket_sales.db') # allow the communication - with arg name of the database
cursor = conn.cursor()

# create table
create_table = """ CREATE TABLE IF NOT EXISTS sales(
                        invoice_id INT,
                        city TEXT,
                        gender TEXT,
                        product_line TEXT,
                        unit_price FLOAT,
                        quantity INT,
                        tax FLOAT,
                        total FLOAT,
                        payment TEXT,
                        gross_margin_percentage FLOAT,
                        gross_income FLOAT
                        ); """

cursor.execute(create_table)

# insert data from csv file
with open(r"C:\Users\Gaurav Malik\OneDrive\Documents\DAB111\DAB111_project_Group_2-main\DAB111_project_Group_2-main\supermarket_sales.csv") as f:
    data = csv.reader(f)
    for row in data:
        if row[0]=='ID': 
            continue
        cursor.execute("INSERT INTO sales VALUES (?, ?, ?, ?, ?, ?, ? ,? ,? ,?, ?)", (row[0], row[1], row[2], row[3], row[4], row[5], row[6], row[7], row[8], row[9], row[10]))

results = cursor.execute("SELECT * FROM sales;").fetchall()

# save data
conn.commit()
#conn.close()
```
